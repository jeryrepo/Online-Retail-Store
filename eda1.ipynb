# EDA

# Total number of customers.

```python
num_customers = df["Customer ID"].nunique()
```

```python
fig, ax = plt.subplots()
ax.text(0.5, 0.5, f'Total Number Of Cutomers: {num_customers} ', transform=ax.transAxes, fontsize=24, ha='center')
ax.axis('off')
plt.show()
```

# Find avg price by country.

```python
average_prices_by_country = df.groupby('Country')['Price'].mean()
print("average_prices_by_country:")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(average_prices_by_country.index, average_prices_by_country.values)
plt.xlabel('Country')
plt.ylabel('Average Price')
plt.title('Average Price by Country')
plt.xticks(rotation=90)
plt.show()
```

# What is the distribution of customers by country?

```python
customer_distribution = df.groupby('Country')['Customer ID'].nunique().reset_index()
customer_distribution = customer_distribution.sort_values(by='Customer ID', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(customer_distribution['Country'], customer_distribution['Customer ID'])
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.title('Distribution of Customers by Country')
plt.xticks(rotation=90)
plt.show()
```

# Most sellable & least sellable// identify unusual patterns in the quantities of products purchased.

```python
product_sales = df.groupby('Description')['Quantity'].sum().reset_index()
most_popular_products = product_sales.sort_values(by='Quantity', ascending=False).head(5)
least_popular_products = product_sales.sort_values(by='Quantity', ascending=True).head(5)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

ax1.bar(most_popular_products['Description'], most_popular_products['Quantity'])
ax1.set_xlabel('Product Description')
ax1.set_ylabel('Quantity Sold')
ax1.set_title('Most Sellable Products')
ax1.tick_params(axis='x', rotation=90)

ax2.bar(least_popular_products['Description'], least_popular_products['Quantity'])
ax2.set_xlabel('Product Description')
ax2.set_ylabel('Quantity Sold')
ax2.set_title('Least Sellable Products')
ax2.tick_params(axis='x', rotation=90)
plt.show()
```
